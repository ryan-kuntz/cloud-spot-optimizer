# Feature Engineering

Documents the feature decisions made for this project.

**Philosophy:** EDA reveals what exists in the data. Feature engineering decides how to represent it for the model.

The actual feature engineering code lives in `src/features.py`. This notebook explains the reasoning behind each decision.

---

## Features currently in use

| Feature | Type | Rationale |
|---|---|---|
| `hour` | Time | Spot prices vary by time of day — demand follows business hours |
| `day_of_week` | Time | Weekday vs weekend demand patterns (validated in EDA) |
| `month` | Time | Seasonal patterns — limited signal with one month of data |
| `is_weekend` | Time | Binary encoding of weekend — complements day_of_week |
| `natgas_price` | Macro | Energy cost proxy for data center operating costs |

---

## Features to add with more data

The following require multiple months of data to be meaningful:

| Feature | Requires | Rationale |
|---|---|---|
| `price_lag_1h` | Multi-month data | Most recent price is the strongest predictor |
| `price_lag_24h` | Multi-month data | Same hour yesterday |
| `price_lag_7d` | Multi-month data | Same hour last week — captures weekly cycles |
| `price_roll_mean_24h` | Multi-month data | Recent price trend |
| `price_roll_std_24h` | Multi-month data | Recent price volatility |

---

## Features to consider adding in future

| Feature | Reason not used |
|---|---|
| AWS service health incidents | Requires scraping AWS health dashboard |
| Big Tech earnings dates | Interesting signal but hard to quantify |
| DXY (USD strength) | Low priority relative to time and lag features |
| VIX (market volatility) | Indirect signal — deprioritized for now |
